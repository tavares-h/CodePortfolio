```python
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
import statsmodels.api as sm


# Load the diabetes data set.
#diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
df = pd.read_csv('california_housing_train.csv', index_col = 0)
predictor1 = df.population
predictor2 = df.households
predictor3 = df.total_rooms
predictor4 = df.total_bedrooms
predictor5 = df.housing_median_age
predictor6 = df.median_income

response = df.median_house_value
predictor1 = predictor1.values.reshape(-1, 1)
predictor2 = predictor2.values.reshape(-1, 1)
predictor3 = predictor3.values.reshape(-1, 1)
predictor4 = predictor4.values.reshape(-1, 1)
predictor5 = predictor5.values.reshape(-1, 1)
predictor6 = predictor6.values.reshape(-1, 1)
response = response.values.reshape(-1, 1)

print("Correlation between Population and Median House Value is :", np.corrcoef(df.population, df.median_house_value) [0,1])
print("\n")

print("Correlation between Households and Median House Value is :", np.corrcoef(df.households, df.median_house_value) [0,1])
print("\n")

print("Correlation between Total Rooms and Median House Value is :", np.corrcoef(df.total_rooms, df.median_house_value) [0,1])
print("\n")

print("Correlation between Median Income and Median House Value is :", np.corrcoef(df.total_bedrooms, df.median_house_value) [0,1])
print("\n")

print("Correlation between Median Income and Median House Value is :", np.corrcoef(df.housing_median_age, df.median_house_value) [0,1])
print("\n")

print("Correlation between Median Income and Median House Value is :", np.corrcoef(df.median_income, df.median_house_value) [0,1])
print("\n")

#print("Only Median Income vs. Median House Value")
#print("Linear")
#regr_linear = linear_model.LinearRegression()
#regr_linear.fit(predictor6, response)
#print('Coef', regr_linear.coef_, 'Intercept', regr_linear.intercept_)

#print("Ridge")
#regr_Ridge = linear_model.Ridge(alpha=0.1)
#regr_Ridge.fit(predictor6, response)
#print('Coef', regr_Ridge.coef_, 'Intercept', regr_Ridge.intercept_)

print("\nMedian Income vs Median House Value (unregularized)")
onepredictors = df.median_income
onepredictors = sm.add_constant(onepredictors)
lr_model = sm.OLS(response,onepredictors).fit()
print(lr_model.summary())
print(lr_model.params)

print("\nMedian Income vs Median House Value (regularized)")
onepredictors = df.median_income
onepredictors = sm.add_constant(onepredictors)
lr_model = sm.OLS(response,onepredictors).fit_regularized(method="elastic_net")
#print(lr_model.summary())
print(lr_model.params)

print("\nMedian Housing Age vs Median House Value (unregularized)")
onepredictors = df.housing_median_age
onepredictors = sm.add_constant(onepredictors)
lr_model = sm.OLS(response,onepredictors).fit()
print(lr_model.summary())
print(lr_model.params)

print("\nMedian Housing Age vs Median House Value (regularized)")
onepredictors = df.housing_median_age
onepredictors = sm.add_constant(onepredictors)
lr_model = sm.OLS(response,onepredictors).fit_regularized(method="elastic_net")
#print(lr_model.summary())
print(lr_model.params)

print("\nMedian Income & Median Houseing Age vs Median House Value (unregularized)")
twopredictors = df[["median_income","housing_median_age"]]
twopredictors = sm.add_constant(twopredictors)
lr_model = sm.OLS(response,twopredictors).fit()
print(lr_model.summary())
print(lr_model.params)

print("\nMedian Income & Median Houseing Age vs Median House Value (regularized)")
twopredictors = df[["median_income","housing_median_age"]]
twopredictors = sm.add_constant(twopredictors)
lr_model = sm.OLS(response,twopredictors).fit_regularized(method="elastic_net")
#print(lr_model.summary())
print(lr_model.params)

print("\nMedian Income & Median Houseing Age & Total Rooms vs Median House Value (unregularized)")
threepredictors = df[["median_income","housing_median_age","total_rooms"]]
threepredictors = sm.add_constant(threepredictors)
lr_model = sm.OLS(response,threepredictors).fit()
print(lr_model.summary())
print(lr_model.params)

print("\nMedian Income & Median Houseing Age & Total Rooms vs Median House Value (regularized)")
threepredictors = df[["median_income","housing_median_age","total_rooms"]]
threepredictors = sm.add_constant(threepredictors)
lr_model = sm.OLS(response,threepredictors).fit_regularized(method="elastic_net")
#print(lr_model.summary())
print(lr_model.params)
```
